```python
from sklearn.datasets import load_digits  # load digit is one of the dataset which is stored in sklearn for practice
%matplotlib inline
import matplotlib.pyplot as plt
digits = load_digits()             # storing the dataset
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
from sklearn.model_selection import train_test_split
X = digits.data
y = digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # splitting
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.svm import SVC
model = SVC()                  # building model

model.fit(X_train, y_train)    # training model 
```

```python
model.score(X_test, y_test) # 99.44% accuracy! Looks like optimal
```

Let's try to tune the parameter of model

```python
model_C = SVC(C=10000)     # giving parameter to change default parameter
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test) # getting same score
```

```python
model_g = SVC(gamma=5)    # Chnaging default gamma 
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test) @ Accuracy decreased
```

```python
model_linear_kernal = SVC(kernel='linear') # changing kernel rbf to linear
model_linear_kernal.fit(X_train, y_train)
model_linear_kernal.score(X_test, y_test) # 99.16%
```
